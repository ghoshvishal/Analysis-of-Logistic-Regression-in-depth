# Dropping unwanted columns

```python
df_new=df.drop(columns=['ID', 'Photo', 'Nationality', 'Flag', 'Club', 'Club Logo', 'Work Rate',
       'Body Type', 'Jersey Number', 'Loaned From', 'Crossing', 'Finishing',
       'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling', 'Curve',
       'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning','Potential','Preferred Foot','Skill Moves', 'Weak Foot','GKReflexes','Position'])
df_new.head()
```

```python
df_new.info()
```

# Value and Wage column - stripping signs and cnverting into float

```python
df_new['Value']= df_new['Value'].map(lambda x: x.lstrip('€').rstrip('KM'))
# df_new['Value']= df_new['Value'].map()
df_new['Wage']= df_new['Wage'].map(lambda x: x.lstrip('€').rstrip('KM'))
# df_new['Wage']= df_new['Wage'].map(lambda x: x)
df_new.head()
```

```python
df_new['Value']=df_new['Value'].astype(dtype= 'float',inplace=True)
df_new['Wage']=df_new['Wage'].astype(dtype='float',inplace =True)
df_new.info()
```

```python
df_new.head()
```

# joined column - get only year and Convert dtype to int

```python
df_new['Joined'] = pd.to_datetime(df_new.Joined)

df_new['Joined']= pd.DatetimeIndex(df_new['Joined']).year
df_new['Joined'].head()
```

```python
df_new.info()
```

```python
df_new.head()
```

```python
df_new['Joined']=df_new.Joined.fillna(df_new.Joined.mean()).astype(dtype='int')
```

```python
df_new.Joined.head()
```

```python
df_new
```

# contract valid until column : Convert str to datetime

```python
df_new['Contract Valid Until'].astype(dtype='datetime', error = 'ignore')
```

#  Weight column : removing lbs and converting it into float

```python
df_new['Weight']=df_new.Weight.astype(dtype='str').map(lambda x: x.rstrip('lbs'))
```

```python
df_new['Weight'].head()
```

```python
df_new.head()
```

# Release Clause : Amount with Euro symbol as prefix and suffix ‘K’ or ‘M’ indicating thousands and millions respectively; convert it into float

```python
df_new['Release Clause']=df_new['Release Clause'].astype(dtype='str').map(lambda x: x.lstrip('€').rstrip('KM')).astype(dtype='float')
```
